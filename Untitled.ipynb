```python
%matplotlib inline

import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('data/data.csv')
df.loc[df['Do you use a banking application'] == 'Yes']
df = df.dropna(how='any',axis=0)
```

```python
df.head()
```

```python
avr_age = {
    'below 18' : 15,
    '18-25' : 23,
    '25-30' : 28,
    '30-35' : 33,
    '35-45' : 40,
    '45-60' : 52,
    'above 60' : 67
    
}
avr_income = {
    'less than 20000': 15000,
    'between 20000 - 50000': 35000,
    'between 50000 - 100000': 75000,
    'between 100000 - 250000': 175000,
    'above 250000': 350000,
    'Not Applicable': np.nan
}

avg_visits = {
    'daily':365,
    'once a week':52,
    'once a month':12,
    'once every few months':4,
    'never': np.nan   
}
transaction_avg = {
    'below 5000': 3000,
    '5000 -10000': 7500,
    '10000 -25000': 17500,
    '25000 - 50000':37500,
    '50000 - 100000':75000,
    '100000 - 200000':150000,
    'above 200000': 300000,
    'N/A': np.nan,
    'nan':np.nan
}
yes_no = {
    'Yes': 1,
    'No': 0,
    'Maybe': 0.5
}
```

```python
for i, row in df.iterrows():
    df.set_value(i,'age',avr_age[row['How old are you']])
    df.set_value(i,'income',avr_income[row['Monthly income']])
    df.set_value(i,'visits',avg_visits[row['how often do you visit the bank']])
    df.set_value(i,'transaction',transaction_avg[row['transactions through the application monthly']])
#     df.set_value(i,'secure',yes_no[row['do you feel your app is secure']])
    df.set_value(i,'useful',yes_no[row['Are you satisfied with the app']])
    df.set_value(i,'complex',yes_no[row['is you application very complex']])
```

```python
for i, row in df.iterrows():
    df.set_value(i,'spending', min(row['transaction']/row['income']*100,100))
```
